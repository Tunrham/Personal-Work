```python
## Below shows the breakdown of the age of all 90 People in the Sample
### As we can see, the data is significantly skewed (right-tailed) and majority of the
### Data falls in the 18-30 year old range.
#### This means that samples in the 50+ age range can skew the dataset - 
#### especially when using things such as "Mean" or "Average"
sns.distplot(ds['age'], norm_hist=True)

##### Fun fact the Age Column is what's known as "Continuous" Data - Or Data that can lie on an "infinite" range
```

```python
## Here's a distribution of the sample data based on gender
### As we can see, this dataset is Heavily female, and, this could 
### heavily influence results. This is known as "Data Imbalance"
sns.countplot(ds['gender'])
```

```python
## This is a violin plot (a Seaborn Python Tool), it shows the frequency of Age of Samples in each Gender
### Given there are only like 15 males in the whole dataset, the Male violin is a bit less smooth 
### as it represents a much smaller dataset
sns.violinplot(x='gender', y='age', data= ds)
```
