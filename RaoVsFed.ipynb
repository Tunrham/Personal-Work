```python
# Imports

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# Bring in our dataset:

ds = pd.read_csv("RaoVsFed.csv")

ds.head()
ds.info()
```

```python
# Let's Breakdown each player's serve based on side of the court

## Federer Duece Side
rf_ds = ds[(ds["CourtSide"]== "Due") & (ds["Serving"]=="RF")]
print("There are: ", len(rf_ds), " Point of Federer Serving Duece Side")
## Federer Ad Side
rf_as = ds[(ds["CourtSide"]== "Adv") & (ds["Serving"]=="RF")]
print("There are: ", len(rf_as), " Point of Federer Serving Ad Side")
## Raonic Duece Side
mr_ds = ds[(ds["CourtSide"]== "Due") & (ds["Serving"]=="MR")]
print("There are: ", len(mr_ds), " Point of Raonic Serving Duece Side")
## Raonic Ad Side
mr_as = ds[(ds["CourtSide"]== "Adv") & (ds["Serving"]=="MR")]
print("There are: ", len(mr_ds), " Point of Raonic Serving Ad Side")
```

```python
# Let's look at some generic stuff, such as:

## Rally Length by Side and Server
sns.kdeplot(rf_ds['rallyCount'], label = "Federer Serve Duece Side P Length")
sns.kdeplot(rf_as['rallyCount'], label = "Federer Ad Duece Side P Length")
sns.kdeplot(mr_ds['rallyCount'], label = "Raonic Serve Duece Side P Length")
sns.kdeplot(mr_as['rallyCount'], label = "Raonic Ad Duece Side P Length")


### Two good servers it's fairly common to see short points (6 shots and longer is uncommon)
```

```python
# Sure, let's get more specific, say; Federer Service Frequencies by Side

## Federer Duece Side
print("-----------------Federer-------------------")
print("------------Federer-Duece-Side-------------")
print("-------------------------------------------")
print("----------------1st Serves-----------------")
x = rf_ds['T1st'].str[0].value_counts() # Here's a Dataframe for 1st Serves (only) Duece Side
print("Federer Hit: ", x['6'], " of ", x.sum(), " 1st serves Tee")
print("Federer Hit: ", x['5'], " of ", x.sum(), " 1st serves Body")
print("Federer Hit: ", x['4'], " of ", x.sum(), " 1st serves Wide")
print("-------------------------------------------")
print("-------------------------------------------")
print("Tee represents: ", round((x['6'])/(x.sum()),2),"%", " of his 1st serves Duece Side")
print("Body represents: ", round((x['5'])/(x.sum()),2),"%", " of his 1st serves Duece Side")
print("Wide represents: ", round((x['4'])/(x.sum()),2),"%", " of his 1st serves Duece Side")
print("-------------------------------------------")
print("----------------2nd Serves-----------------")
y = rf_ds['T2nd'].str[0].value_counts() # here's a Dataframe from 2nd Serves (only) Duece Side
print("Federer Hit: ", y['6'], " of ", y.sum(), " 2nd serves Tee")
print("Federer Hit: ", y['5'], " of ", y.sum(), " 2nd serves Body")
print("Federer Hit: ", y['4'], " of ", y.sum(), " 2nd serves Wide")
print("-------------------------------------------")
print("-------------------------------------------")
print("Tee represents: ", round((y['6'])/(y.sum()),2),"%", " of his 2nd serves Duece Side")
print("Body represents: ", round((y['5'])/(y.sum()),2),"%", " of his 2nd serves Duece Side")
print("Wide represents: ", round((y['4'])/(y.sum()),2),"%", " of his 2nd serves Duece Side")
print("-------------------------------------------")
print("------------Federer-Advan-Side-------------")
print("-------------------------------------------")
print("----------------1st Serves-----------------")
q = rf_as['T1st'].str[0].value_counts() # Here's a Dataframe for 1st Serves (only) Advantage Side
print("Federer Hit: ", q['6'], " of ", q.sum(), " 1st serves Tee")
print("Federer Hit: ", q['5'], " of ", q.sum(), " 1st serves Body")
print("Federer Hit: ", q['4'], " of ", q.sum(), " 1st serves Wide")
print("-------------------------------------------")
print("-------------------------------------------")
print("Tee represents: ", round((q['6'])/(q.sum()),2),"%", " of his 1st serves Advantage Side")
print("Body represents: ", round((q['5'])/(q.sum()),2),"%", " of his 1st serves Advantage Side")
print("Wide represents: ", round((q['4'])/(q.sum()),2),"%", " of his 1st serves Advantage Side")
print("-------------------------------------------")
print("----------------2nd Serves-----------------")
a = rf_as['T2nd'].str[0].value_counts() # here's a Dataframe from 2nd Serves (only) Duece Side
print("Federer Hit: ", a['6'], " of ", a.sum(), " 2nd serves Tee")
print("Federer Hit: ", a['5'], " of ", a.sum(), " 2nd serves Body")
print("Federer Hit: ", a['4'], " of ", a.sum(), " 2nd serves Wide")
print("-------------------------------------------")
print("-------------------------------------------")
print("Tee represents: ", round((a['6'])/(a.sum()),2),"%", " of his 2nd serves Advantage Side")
print("Body represents: ", round((a['5'])/(a.sum()),2),"%", " of his 2nd serves Advantage Side")
print("Wide represents: ", round((a['4'])/(a.sum()),2),"%", " of his 2nd serves Advantage Side")
```

```python
# General Notes

# Federer's Second Serves on the Advantage Side seem slightly out of Line.
# Like, in the few matches with players that I've seen, a majority of these
# serves are going wide - it's very possible that Raonic is Camping out
# looking for a forehand on Second serves and that's affecting the data
# by having it labelled as "body Serve"?

## Only one way to find out I guess:

as_return = rf_as[(rf_as["Sv2"]== "5")]
rr = as_return['T2nd'].str[2].value_counts()

print("Raonic hit : ", rr['b'], "Backhands")
print("Raonic hit : ", rr['f'], "Forehands")
print("Raonic hit : ", rr['s'], "Slice Backhands")
print("Raonic hit : ", rr['v'], "Volley", "# This is a Data error I'd hope")


# Well, looks like I was wrong, on a lot of those body serves Raonic is Hitting Backhands
# So it's very possible that Federer is just balanced on his second serves Ad Side
```
