```python
# Import Libraries:

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# Bring in Test Dataset:

ds = pd.read_csv("DjoWaw.csv")

```

```python
# Preprocessing Steps

## Remove Tiebreaks

notb = ds[(ds.TBpt.isna())]


## Made a "Processed CSV" that will be written to current directory
notb.to_csv('DjoWawProcessed.csv')

```

```python
## Add Spaces to the Rally Data on first serves:
def stringTokenizer(csvname):
    row = 0
    col = 14 # This is the First Serve Column
    
    while row<len(csvname):
        string = csvname.iloc[row, col]
    
        string = string.replace("f", " f")
        string = string.replace("b", " b")
        string = string.replace("r", " r")
        string = string.replace("s", " s")
        string = string.replace("v", " v")
        string = string.replace("z", " z")
        string = string.replace("o", " o")
        string = string.replace("p", " p")
        string = string.replace("u", " u")
        string = string.replace("y", " y")
        string = string.replace("l", " l")
        string = string.replace("m", " m")
        string = string.replace("h", " h")
        string = string.replace("i", " i")
        string = string.replace("j", " j")
        string = string.replace("k", " k")
        string = string.replace("t", " t")
        string = string.replace("q", " q")
        
        print(string)
    
        row = row+1
```

```python
# Copy pasted this over to CSV, it's not the best way, but at the moment I can't get .replace() inplace=True working
stringTokenizer(notb)
```
